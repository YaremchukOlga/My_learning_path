Ниже я объединила таблицы sellers и geolocation по общей колонке с кодами продавцов и удалила повторяющиеся значение:

```python
sellers = sellers.set_index('geolocation_zip_code_prefix').join(geolocation.drop_duplicates('geolocation_zip_code_prefix', keep = 'first').set_index('geolocation_zip_code_prefix'), how = 'left')
```

```python
sellers
```

```python
order_items = order_items.set_index('seller_id').join(sellers.set_index('seller_id'), how = 'left').query('order_item_id == 1')
```

```python
result = order_items.join(logistics.set_index('order_id'), on = 'order_id', how = 'left')[['order_id','geolocation_lat', 'geolocation_lng', 'delivery_delay_days']].reset_index()
```

```python
result
```

## Сформируем рейтинг магазинов по потенциалу улучшения

```python
result['delivery_delay_days'] = result['delivery_delay_days'].apply(lambda x: x if x > 0 else 0)
```

```python
result.head()
```

Для каждого магазина посчитаем число заказов и среднее время опоздания:

```python
seller_stats = result.groupby(['seller_id']).aggregate({'order_id':'count',
                                                       'delivery_delay_days': 'mean',
                                                       'geolocation_lat': 'first',
                                                       'geolocation_lng': 'first'})\
                                                       .sort_values(['order_id'], ascending = False)
```

```python
seller_stats.columns = ['заказы','среднее_опоздание', 'geolocation_lat', 'geolocation_lng']
```

```python
seller_stats
```

```python
seller_stats['сколько_опоздал_в_сумме'] = (seller_stats['среднее_опоздание'] * seller_stats['заказы']).round(0)
```

```python
seller_stats.sort_values(['среднее_опоздание'], ascending = False)
```

Посчитаем средний показатель по опозданиям и отсечём магазины с малым числом заказов:

```python
seller_stats.query('заказы > 100')
```

```python
mean = seller_stats.query('заказы > 100')['среднее_опоздание'].mean()
#среднее опоздание среди популярных магазинов
```

```python
seller_stats['среднее_опоздание'].mean()
#среднее опоздание по всем магазинам из данных
```

```python
mean
```

Посчитаем метрику: на сколько дней магазин бы опаздывал в сумме, если бы работал на уровне среднестатистического магазина

```python
seller_stats['мог_бы_опоздать_в_сумме'] = (mean * seller_stats['заказы']).round(0)
```

```python
seller_stats
```

Вычтем дни и получим потенциал - выигрыш в сумме дней опозданий

```python
seller_stats['сэкономленные_дни'] = seller_stats['сколько_опоздал_в_сумме'] - seller_stats['мог_бы_опоздать_в_сумме']
```

```python
#отсортируем магазины по потенциалу улучшения
seller_stats.sort_values(['сэкономленные_дни'], ascending = False, inplace = True)
```

```python
seller_stats[['заказы', 'среднее_опоздание', 'сэкономленные_дни']].head()
```
